```python
AQI_BREAKPOINTS = {
    "PM2.5": [
        (0,30,0,50),
        (31,60,51,100),
        (61,90,101,200),
        (91,120,201,300),
        (121,250,301,400),
        (251,500,401,500)
    ],

    "PM10":[
        (0,50,0,50),
        (51,100,51,100),
        (101,250,101,200),
        (251,350,201,300),
        (351,430,301,400),
        (431,600,401,500)
    ],

    "NO2":[
        (0,40,0,50),
        (41,80,51,100),
        (81,180,101,200),
        (181,280,201,300),
        (281,400,301,400),
        (401,1000,401,500)
    ],

    "SO2":[
        (0,40,0,50),
        (41,80,51,100),
        (81,380,101,200),
        (381,800,201,300),
        (801,1600,301,400),
        (1601,2000,401,500)
    ],

    "CO":[
        (0,1,0,50),
        (1.1,2,51,100),
        (2.1,10,101,200),
        (10.1,17,201,300),
        (17.1,34,301,400),
        (34.1,50,401,500)
    ],

    "Ozone":[
        (0,50,0,50),
        (51,100,51,100),
        (101,168,101,200),
        (169,208,201,300),
        (209,748,301,400),
        (749,1000,401,500)
    ],

    "NH3":[
        (0,200,0,50),
        (201,400,51,100),
        (401,800,101,200),
        (801,1200,201,300),
        (1201,1800,301,400),
        (1801,2000,401,500)
    ]
}
```

```python
def calculate_subindex(value, pollutant):

    if pd.isna(value):
        return np.nan

    breakpoints = AQI_BREAKPOINTS[pollutant]

    for BP_LO, BP_HI, I_LO, I_HI in breakpoints:

        if BP_LO <= value <= BP_HI:

            return round(
                ((I_HI-I_LO)/(BP_HI-BP_LO))*
                (value-BP_LO)+I_LO
            )

    return np.nan
```

```python
df1["PM25_Index"] = df1["PM2.5"].apply(lambda x: calculate_subindex(x,"PM2.5"))

df1["PM10_Index"] = df1["PM10"].apply(lambda x: calculate_subindex(x,"PM10"))

df1["NO2_Index"] = df1["NO2"].apply(lambda x: calculate_subindex(x,"NO2"))

df1["SO2_Index"] = df1["SO2"].apply(lambda x: calculate_subindex(x,"SO2"))

df1["CO_Index"] = df1["CO"].apply(lambda x: calculate_subindex(x,"CO"))

df1["Ozone_Index"] = df1["Ozone"].apply(lambda x: calculate_subindex(x,"Ozone"))

df1["NH3_Index"] = df1["NH3"].apply(lambda x: calculate_subindex(x,"NH3"))
```

```python
df1["AQI"] = df1[
    [
        "PM25_Index",
        "PM10_Index",
        "NO2_Index",
        "SO2_Index",
        "CO_Index",
        "Ozone_Index",
        "NH3_Index"
    ]
].max(axis=1)
```

```python
def aqi_category(aqi):

    if pd.isna(aqi):
        return np.nan

    if aqi <= 50:
        return "Good"

    elif aqi <=100:
        return "Satisfactory"

    elif aqi <=200:
        return "Moderate"

    elif aqi <=300:
        return "Poor"

    elif aqi <=400:
        return "Very Poor"

    else:
        return "Severe"
```

```python
df1["AQI Category"] = df1["AQI"].apply(aqi_category)
```
