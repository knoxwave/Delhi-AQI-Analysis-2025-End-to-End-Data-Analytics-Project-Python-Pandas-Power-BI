```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "Delhi_AQI_Final_Dashboard.csv"
)
```

```python
df.head()
```

```python
df = df[(df["AQI"].notna()) & (df["AQI"] != 0)]
```

```python
# Count AQI = 0
aqi_zero = (df["AQI"] == 0).sum()

# Count AQI is blank (NaN)
aqi_blank = df["AQI"].isna().sum()

# Total invalid AQI
invalid_aqi = aqi_zero + aqi_blank

# Valid AQI
valid_aqi = len(df) - invalid_aqi

# Total rows
total_rows = len(df)

print("="*40)
print(f"Total Rows      : {total_rows:,}")
print(f"AQI = 0         : {aqi_zero:,}")
print(f"AQI Blank (NaN) : {aqi_blank:,}")
print(f"Invalid AQI     : {invalid_aqi:,}")
print(f"Valid AQI       : {valid_aqi:,}")
print("="*40)
```

```python
df.to_csv("Delhi_AQI_Final_Clean1.csv", index=False)
```

```python
df2 = pd.read_csv(
    "Delhi_AQI_Final_Clean1.csv"
)
print(df2.isnull().sum())
```

```python
numeric_cols = df2.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    df2[col] = df2.groupby("Station")[col].transform(
        lambda x: x.fillna(x.median())
    )
```

```python
print(df2.isnull().sum())
```

```python
df2.to_csv("Delhi_AQI_Final_Clean3.csv", index=False)
```
