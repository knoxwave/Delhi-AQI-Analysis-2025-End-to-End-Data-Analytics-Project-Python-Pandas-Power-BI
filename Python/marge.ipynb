```python
import pandas as pd
import glob
import os
import re

# Folder containing all CSV files
folder_path = r"C:\Users\Leo\Downloads\aqi"      # <-- Change this path

# Get all CSV files
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

print(f"Found {len(csv_files)} CSV files.\n")

all_data = []

for file in csv_files:

    # Read CSV
    df = pd.read_csv(file, na_values=["NA"])

    # Get filename
    filename = os.path.basename(file)

    # Extract station name from filename
    # Example:
    # raw_data_hourly_anand_vihar_delhi_-_dpcc_1H.csv
    station = re.sub(r"raw_data_hourly_", "", filename)
    station = re.sub(r"_delhi.*", "", station)
    station = station.replace("_", " ").title()

    # Add Station column
    df["Station"] = station

    all_data.append(df)

    print(f"Loaded: {station} ({len(df)} rows)")

# Merge all files
merged_df = pd.concat(all_data, ignore_index=True)

# Convert Timestamp to datetime
merged_df["Timestamp"] = pd.to_datetime(
    merged_df["Timestamp"],
    format="%d-%m-%Y %H:%M",
    errors="coerce"
)

# Sort by Station and Timestamp
merged_df = merged_df.sort_values(
    ["Station", "Timestamp"]
).reset_index(drop=True)

# Save merged CSV
output_file = os.path.join(folder_path, "Delhi_AQI_2025_Merged.csv")
merged_df.to_csv(output_file, index=False)

print("\n=================================")
print("Merge Completed Successfully")
print("=================================")
print("Total Files :", len(csv_files))
print("Total Rows  :", len(merged_df))
print("Output File :", output_file)
```

```python
import pandas as pd
import glob
import os
import re

# Folder containing all CSV files
folder_path = r"C:\Users\Leo\Downloads\aqi"

# Get all CSV files
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

print(f"Found {len(csv_files)} CSV files.\n")

all_data = []

for file in csv_files:

    # Read CSV
    df = pd.read_csv(file, na_values=["NA"])

    # Remove hidden spaces/BOM from column names
    df.columns = (
        df.columns
        .str.replace("\ufeff", "", regex=False)
        .str.strip()
    )

    # Clean Timestamp column
    if "Timestamp" in df.columns:
        df["Timestamp"] = (
            df["Timestamp"]
            .astype(str)
            .str.strip()
        )

        # Auto detect date format
        df["Timestamp"] = pd.to_datetime(
            df["Timestamp"],
            dayfirst=True,
            errors="coerce"
        )

    # Extract station name
    filename = os.path.basename(file)

    station = re.sub(r"raw_data_hourly_", "", filename)
    station = re.sub(r"_delhi.*", "", station)
    station = station.replace("_", " ").title()

    df["Station"] = station

    all_data.append(df)

    print(f"Loaded: {station} ({len(df)} rows)")

# Merge all files
merged_df = pd.concat(all_data, ignore_index=True)

# Sort
merged_df = merged_df.sort_values(
    ["Station", "Timestamp"]
).reset_index(drop=True)

# Save
output_file = os.path.join(folder_path, "Delhi_AQI_2025_Merged.csv")
merged_df.to_csv(output_file, index=False)

print("Done.")
```
